4. Usamos *df.describe( )* para obtener datos estadísticos de las variables numéricas encontradas en el archivo, en este caso, las relevantes son **mean** (Media), **50%** (Mediana), **std** (Desviación estándar) y aunque no se encuentra en la tabla, **mode** (Moda).

```python
df.describe()
```

5. Creamos una función para determinar estos valores para todas las variables numéricas (Las columnas de la tabla anterior).
- Gracias al uso de las librerías previamente importadas, no debemos especificar la operación para cada medida y solo debemos invocar las palabras reservadas.

```python
def estadisticas(columna):
  media = df[columna].mean()
  mediana = df[columna].median()
  desviacion_estandar= df[columna].std()
  moda=df[columna].mode()[0]
  return print (f'\n Para "{columna}" las medidas de tendencia central son: \n Media:{media:.2f} \t Mediana:{mediana:.2f} \t Desviación estándar:{desviacion_estandar:.2f} \t Moda:{moda:.2f}')
```

6. Definimos el título de los resultados a visualizar, guardamos los nombres de las columnas a utilizar en *variables_numericas* y usamos un ciclo *for* para invocar la función creada en el paso anterior para las columnas almacenadas en la variable.

```python
print ('Medidas de tendencia central para las variables numéricas')
variables_numericas = ['incident_id','n_killed', 'n_injured','congressional_district', 'latitude','longitude','n_guns_involved','state_house_district','state_senate_district']
for columna in variables_numericas:
  estadisticas(columna)
```

7. Con un segundo ciclo for, definimos que para las columnas almacenadas en la variable, se cree un histograma de frecuencias para cada una, usando plt.title para asociar la columna a su respectivo histograma y plt.show para mostrar cada uno de los gráficos en la misma ejecución.

```python
for columna in variables_numericas:
  df[columna].plot.hist()
  plt.title(f'Histograma de "{columna}"')
  plt.show()
```

8. Con un tercer ciclo for definimos que para las columnas almacenadas en la variable, se cree un diagrama de cajas y bigotes para cada una, usando plt.title para asociar la columna a su respectivo diagrama y plt.show para mostrar cada uno de los gráficos en la misma ejecución.

```python
for columna in variables_numericas:
  df[columna].plot.box()
  plt.title(f'Diagrama de cajas y bigotes de "{columna}"')
  plt.show()
```
